```
# Use hashtags for comments
# Always explain what you are writing as you go so that other people can read and understand your code 

# Lets start with some basic python 
print("Python Variables")

# Printing is powerful in Python. you can print any number of variables with differing data types

# Variables can be created and initialised without types. Python will infer the type from the assignment. 
x = 1 # Integer
print(x, type(x))

# We can change variables and their types on the go.
x = "Hello World" # String
print(x, type(x))

x = 3.14 # Float (non-integer)
print(x, type(x))

# Some Built-ins and Print
x = "Hello World"

print()
print("String Length: ", len(x))
print("String Type: ", type(x))
print("Split Words", x.split())

# In this section, we have only used a single variable "x". However we can have any number of variables that we want. 

# Python's print statement will print almost all python objects. Use it extensively to try to understand and debug what is going on. 
```

# Python Lists

- same idea as arrays

```
# Python Lists 
x = "Hello World"

print("Python Lists")

# We can store different kinds of variables in a list:
my_list = [1, 2, "Test", 2.35] # Create a list
print("My List: ", my_list)

x_list = list(x)  # Convert string to a list
print("String List: ", x_list)
```

# User Input 

```
# Python User Interactivity 

number = input("Please enter a number: ")
print("The number you have entered is",number)

# we can convert data types. Basic data types that you need to know are strings, integers and floats

print("the number you entered is currently a ",type(number))

number_converted = int(number)

print("Now the number you entered is a ", type(number_converted) )
```

# Control Flow 

```
# In python you can loop through almost any iterable container (list, dictionary, etc). Indexes start from 0. 
print("Control Flow")

# Loop through a range of numbers
print("\nNumbers List:")
for i in range(number_converted): 
    print(i)

# Loop through the list. Python will take care of indexing the loop. 
print("\nMy List: ")
for item in my_list:
    print(item)


print("\nString Example:")
# A string is a list of characters in python, so we can loop through it. We can also get the index of the loop using enumerate. 
for i, letter in enumerate(x):
    print(i, letter)
```

# Python Indexing

```
# Indexing Lists and Containers

print("Indexing")
# We can index any part of a list with [] notation:
first_letter = x[0]             # Single Letter
first_five_letters = x[:5]      # Letters 0-4

print("First Letter:", first_letter)
print("First Five: ", first_five_letters)
```
